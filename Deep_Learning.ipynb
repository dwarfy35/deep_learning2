```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load('character_font.npz')
```

```python
print(len(data["images"]))
print(len(data["labels"]))
fig = plt.figure
plt.imshow(data["images"][0], cmap='gray')
plt.show()
```

```python
num = 26*2
images = data["images"][:num]
labels = data["labels"][:num]
```

```python
labels[3]
chr(3)
```

```python
def number_to_letter(number):
    return chr(number + ord('a'))
```

```python
num_row = 4
num_col = 13

# plot images
fig, axes = plt.subplots(num_row, num_col, figsize=(1.5*num_col,2*num_row))
for i in range(num):
    ax = axes[i//num_col, i%num_col]
    ax.imshow(images[i], cmap='gray')
    ax.set_title(number_to_letter(labels[i]))
plt.tight_layout()
plt.show()
```
